```python
import pandas as pd
df=pd.read_csv('dataset/salaries.csv')
df.head()
```

```python
inputs=df.drop('salary_more_than_100k',axis=1)
target=df['salary_more_than_100k']
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), ['company','job','degree'])], remainder='passthrough')
x=columnTransformer.fit_transform(inputs)
```

```python
df=pd.DataFrame(x,columns=['company_1','company_2','company_3','job_1','job_2','job_3','degree_1','degree_2'])
inputs=pd.concat([inputs,df],axis=1)
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree','company_1','job_1','degree_1'],axis=1)
```

```python
inputs_n
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
model=tree.DecisionTreeClassifier()
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.3)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
